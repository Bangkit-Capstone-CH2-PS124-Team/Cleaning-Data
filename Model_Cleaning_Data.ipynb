```python
# Loop through each column name
for col in df.columns:
    # Calculating the percentage of missing values in each column
    value = 100 * (df[col].isnull().sum() / df[col].shape[0])

    # Craft some message to display in the print statement
    msg = 'column:{:>10}\t Percent of NaN value: {:.2f}%'.format(col, value)

    # Print the message
    print(msg)
```

```python
def missing_data_report(data):
    # Calculate the total number of missing values
    total = data.isnull().sum()

    # Calculate the percentage of missing values
    pct = (data.isnull().sum() / data.isnull().count()) * 100

    # Concatenate total and percentage into a DataFrame
    tt = pd.concat([total, pct], axis=1, keys=['Total', 'Percent'])

    # Create a list to store data types
    data_types = []

    # Loop through columns to get data types
    for col in data.columns:
        data_type = str(data[col].dtype)
        data_types.append(data_type)

    # Add the 'Types' column to the report
    tt['Types'] = data_types

    # Transpose the DataFrame for a better view
    return np.transpose(tt)

# Call the function with your DataFrame
missing_data_report(df)
```
